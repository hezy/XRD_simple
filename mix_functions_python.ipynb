```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f1(x, a):
    return a * x**2
```

```python
def f2(x, a):
    return a * x**3 
```

```python
def mix_fun(first_fun, second_fun, n):
    return n * first_fun + (1 - n) * second_fun 
```

```python
x = np.arange (-3, 3 , 1e-2)
```

```python
plt.plot(x, f1(x, 1))
plt.plot(x, f2(x, 1))
plt.plot(x, mix_fun(f1(x, 1), f2(x, 1), 0.5))
```
