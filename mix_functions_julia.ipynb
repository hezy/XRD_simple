```julia
using Plots
```

```julia
f1(x, a) = a * x^2
```

```julia
f2(x, a) = a * x^3
```

```julia
mix_fun(first_fun, second_fun, n) = n * first_fun + (1 - n) * second_fun
```

```julia
x = range(start = -3, stop = 3, step = 0.01)
```

```julia
plot(x, f1.(x, 1))
plot!(x, f2.(x, 1))
plot!(x, mix_fun(f1.(x, 1), f2.(x, 1), 0.5))
```
